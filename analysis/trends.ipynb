```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data/trends/mlops.csv")
```

```python

# convert date column to datetime
df['date'] = pd.to_datetime(df['week'])

# set the figure size and aspect ratio
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

# set the ticks on x-axis
num_ticks = 5
date_range = pd.date_range(start=df['date'].iloc[0], end=df['date'].iloc[-1], periods=num_ticks)
xticks = [date_range[i].strftime('%Y-%m-%d') for i in range(num_ticks)]
ax.set_xticks(date_range)
ax.set_xticklabels(xticks)

# plot the data
ax.plot(df['date'], df['value'])

# add title and labels to the plot
ax.set_title('MLOps term relative popularity')
ax.set_xlabel('Date')
ax.set_ylabel('Relative popularity [%]')

# adjust the font size of the title and labels
ax.title.set_fontsize(20)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)

# set tick font size
ax.tick_params(axis='both', which='major', labelsize=15)
ax.tick_params(axis='both', which='minor', labelsize=15)

# enable grid
ax.grid(visible=True, axis="y")

plt.savefig("output/mlops-popularity.pdf")
```
